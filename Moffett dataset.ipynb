# Seperate categorical and numerical columns and prepare categorical input variables using label encoder

```python
cols =filled_data.columns
num_cols= filled_data._get_numeric_data().columns
cat_cols = list(set(cols) - set(num_cols))
```

```python
from sklearn.preprocessing import LabelEncoder
data1 = filled_data[cat_cols].apply(LabelEncoder().fit_transform)
```

```python
final_data = data1.join(filled_data[num_cols])
```

```python
plt.figure(figsize=(20,16))
sns.heatmap(final_data.corr(),square=True,annot=True,cmap= 'hsv')
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X = final_data.drop('y',axis=1)
y = final_data['y']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=42)
```

```python
classifier= RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=None, random_state=42,max_leaf_nodes=None)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print('confusion Matrix:\n',confusion_matrix(y_pred,y_test))
print("Acurracy  : ",accuracy_score(y_pred,y_test))
```

```python
from sklearn.metrics import f1_score
```

```python
score = f1_score(y_pred, y_test)
print(score)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
X2 = final_data.drop('y',axis=1)
y2 = final_data['y']
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,test_size = 0.3,random_state=42)
```

```python
lr= LogisticRegression(max_iter=10000)
lr.fit(X2_train,y2_train)
y2_pred = lr.predict(X2_test)
```

```python
print('confusion Matrix:\n',confusion_matrix(y2_pred,y2_test))
print("Acurracy  : ",accuracy_score(y2_pred,y_test))
```

# Xgboost classifier

```python
import xgboost
```

```python
X3 = final_data.drop('y',axis=1)
y3 = final_data['y']
X3_train,X3_test,y3_train,y3_test = train_test_split(X3,y3,test_size = 0.3,random_state=42)
```

```python
xgb = xgboost.XGBClassifier(n_estimators=100, learning_rate=0.09, gamma=0, subsample=0.75,colsample_bytree=1, max_depth=None,random_state=42)
xgb.fit(X3_train,y3_train.squeeze().values)
y3_pred = xgb.predict(X3_train)
y3_test_pred = xgb.predict(X3_test)
print(xgb)
```

```python
print('confusion Matrix:\n',confusion_matrix(y3_test_pred,y3_test))
print("Acurracy  : ",accuracy_score(y3_test, y3_test_pred))
```
